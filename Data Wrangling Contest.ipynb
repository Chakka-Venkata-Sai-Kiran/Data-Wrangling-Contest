```python
#Importing the libraries
import pandas as pd
import numpy as np
```

```python
#Importing datasets
df_flight =  pd.read_csv('flights.csv',skiprows=54)
df_planes  =  pd.read_csv('planes.csv')
df_airplanes = pd.read_csv('airlines.csv', skiprows=25)
df_weather =  pd.read_csv('weather.csv',skiprows=42)
```

<B>5.SELECT MIN(year),AVG(year), MAX(year), engine,manufacturer FROM planes GROUP BY engine, manufacturer<B>

```python
#QUERY_05_SOLUTION
query_5=df_planes.groupby(['engine','manufacturer']).agg(m1=('year', min),m2=('year',np.average),m3=('year', max)).rename(columns = {'m1':'MIN(YEAR)','m2':'AVG(YEAR)','m3':'MAX(YEAR)'}).reset_index().iloc[:,[2,3,4,0,1]]
query_5
```

<B>17.SELECT
flights2.
*, 
atemp, 
ahumid
FROM (
SELECT * FROM flights WHERE origin='EWR'
) AS flights2
LEFT
JOIN (
SELECT
year, month, day, 
AVG(temp) AS atemp, 
AVG(humid) AS ahumid
FROM weather
WHERE origin='EWR'
GROUP BY year, month, day<B>

```python
#QUERY_17_SOLUTION
query_17 = pd.merge(df_flight[df_flight.apply(lambda x:x.origin == 'EWR',axis = 1)],df_weather[df_weather.apply(lambda x: x.origin == 'EWR',axis = 1)].groupby(['year','month','day']).agg(m1=('temp', np.average),m2 = ('humid',np.average)).rename(columns = {'m1':'atemp','m2':'ahumid'}).reset_index(),how='left')
query_17
```
